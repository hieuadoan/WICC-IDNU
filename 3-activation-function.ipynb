```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-10, 10, steps=100)
print(x)
```

```python
# Sigmoid activation function
#y = torch.sigmoid(x)
y = 1/(1 + torch.exp(-x))
plt.plot(x, y)
```

```python
# Tanh activation function
y = torch.tanh(x)

plt.plot(x, y)
```

```python
# ReLU activation function
#y = torch.relu(x)
y = torch.max(torch.tensor(0), x)
plt.plot(x, y)
```

```python
# Leaky ReLU activation function
#y = torch.nn.functional.leaky_relu(x, 0.1)
y = torch.max(torch.tensor(0.1)*x, x)
plt.plot(x, y)
```

```python
# Softmax activation function
x = torch.tensor([1.0, 2.0, 3.0])
y = torch.nn.functional.softmax(x, dim=0)
print(y.numpy())
print([(np.exp(i)/sum(np.exp(x))).item() for i in x])
```
